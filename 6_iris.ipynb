## **붓꽃 품종 분류 실습**

```python
import tensorflow as tf
import pandas as pd
```

```python
file_path = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
iris = pd.read_csv(file_path)
iris.head()
```

```python
encoding_iris = pd.get_dummies(iris)
encoding_iris.head()
```

```python
print(encoding_iris.columns)
```

```python
indep_v = encoding_iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
dep_v = encoding_iris[['품종_setosa', '품종_versicolor','품종_virginica']]
print(indep_v.shape, dep_v.shape)
```

# 모델 구조 만들기

```python
X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation='softmax')(X)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy',metrics='accuracy')
```

```python
model.fit(indep_v,dep_v,epochs=100)
```

```python
model.fit(indep_v,dep_v,epochs=100)
```

```python
model.fit(indep_v,dep_v,epochs=100,verbose=0)
```

# 모델을 이용

```python
print(model.predict(indep_v[0:5]))
```

```python
print(dep_v[0:5])
```

```python
print(model.predict(indep_v[-5:]))
print(dep_v[-5:])
```

```python
print(model.get_weights())
```
